```python
import pandas 
import numpy as np
import math 
```

```python
df = pandas.read_csv('data/20192020data.csv')
aggdict = dict()
for item in df.items():
    #print(item[0])
    if item[0] == 'Rk' or item[0] == 'S%' or item[0] == 'FO%':
        print("ok")
        aggdict[item[0]] = 'mean'
    else:
        aggdict[item[0]] = 'sum'
print(aggdict)
clean = df.groupby(['Player']).agg(aggdict) 
stored = 0

lastnames = []
firstnames = []
for i in clean.iterrows():
    #print(i[0])
    first = ""
    last = ""
    f = True
    for letter in i[0]:
        if letter == "\\" :
            break
        if letter == " ":
            f = False
            continue
        if f:
            first = first +letter
        else:
            last += letter
    #print(last,first)
    lastnames.append(last)
    firstnames.append(first)
clean['First'] = firstnames
clean['Last'] = lastnames

clean = clean.sort_values(['Last'])
clean = clean.drop('Rk', 1)
clean
```
