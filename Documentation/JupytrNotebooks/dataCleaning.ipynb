```python
import pandas 
import numpy as np
import math 
```

```python
##player stats
abrvs = ['BOS','WSH','TBL','PHI','PIT','CAR','TOR','CBJ','NYI','NYR','FLA','MTL','BUF','NJD','OTT','DET','STL','VEG','COL','EDM','DAL','WPG','CGY','VAN','NSH','MIN','ARI','CHI','ANA','LAK','SJS']

df = pandas.read_csv('../data/20192020data.csv')
print("NORMAL SIZE:", df.shape)
dfadv = pandas.read_csv('../data/advanced20192020.csv')
print("ADVANCED SIDE:", dfadv.shape)
dfadvtm = pandas.read_csv('../data/advancedteam20192020.csv')
print("ADVANCED TEAM SIZE:", dfadvtm.shape)
dfadv = dfadv.drop(columns=['Rk', 'Age', 'Pos', 'GP', 'TOI/60'])
#dfadv.sort_values(by=['Player'])
dfadvtm = dfadvtm.drop(columns=['S%', 'SV%'])
#for column in dfadv.columns:
#    df[column] = dfadv[column]
#pandas.concat([df,dfadv], axis=1)
dfteams = pandas.read_csv('../data/teams20192020.csv',delimiter = '\t')
print("TEAM SIZE:", dfteams.shape)
for column in dfadvtm.columns:
    dfteams[column] = dfadvtm[column]
aggdict = dict()
for item in df.items():
    #print(item[0])
    if item[0] == 'Rk' or item[0] == 'S%' or item[0] == 'FO%':
        #print("ok")
        aggdict[item[0]] = 'mean'
    else:
        aggdict[item[0]] = 'sum'
clean = df.groupby(['Player']).agg(aggdict) 

#clean = df
stored = 0

lastnames = []
firstnames = []

for i in clean.iterrows():
    #print(i[0])
    first = ""
    last = ""
    f = True
    for letter in i[0]:
        if letter == "\\" :
            break
        if letter == " ":
            f = False
            continue
        if f:
            first = first +letter
        else:
            last += letter
    #print(last,first)
    lastnames.append(last)
    firstnames.append(first)
clean['First'] = firstnames
clean['Last'] = lastnames

#clean = clean.sort_values(['Last'])
clean = clean.drop('Rk', 1)
#clean.columns.name = df.index.name
#clean.index.name = None


#clean = clean.drop('Player', 1)
a = clean['Tm'].unique().tolist()
A = [abv for abv in a if len(abv)<=3]
#print(A)
## RIGHT NOW I STILL NEED TO Adress the Team and the position being weird, but that should be easy
#clean.head(50)
clean = clean[clean['Tm'].isin(abrvs)]
dfadv = dfadv[dfadv['Tm'].isin(abrvs)]
dfadv = dfadv.sort_values(by=['Player'])
dfadv = dfadv.drop(columns=['Player', 'Tm'])
dfadv.reset_index(drop=True, inplace=True)
clean.reset_index(drop=True, inplace=True)
clean = pandas.concat([clean, dfadv], axis=1)
clean
```
